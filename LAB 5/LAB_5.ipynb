```python
#Task 3
#class code
class Spaceship:
  def __init__(self,name,capacity):
    self.__name = name
    self.__capacity = capacity
    self.cargo_details = []
    self.current_weight = 0
    self.cargos = []

  def load_cargo(self,cargo):
    self.cargo_details.append(cargo)
    self.current_weight += cargo.weight
    self.cargos.append(cargo.material)
    if self.current_weight>self.__capacity:
      print(f"Warning: Unable to load Neutronium inside Falcon. Exceeds capacity by {self.current_weight-self.__capacity}.")
      self.current_weight -= cargo.weight
  def display_details(self):
    print(f"Spaceship Name: {self.__name}")
    print(f"Capacity: {self.__capacity}")
    print(f"Current Cargo Weight: {self.current_weight}")
    print(f"Cargo: {self.cargos}")
class Cargo:
  def __init__(self,material,weight):
    self.material = material
    self.weight = weight

#Driver Code
# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
#Task 4
#Class Code
class Student:
  def __init__(self,name,ID,department,email=None,password=None,login_status = False):
    self.name = name
    self.ID = ID
    self.department = department
    self.email = email
    self.password = password
    self.login_status = login_status
    self.courses = 0
    print(f"Student object is created!")
class Usis:
  def __init__(self):
    print(f"USIS is ready to use.")
  def login(self,student):
    if student.email is None and student.password is None:
      print(f"Email and password need to be set.")
    else:
      print(f"Login successful!")
      student.login_status = True
  def advising(self,student,*courses):
    if student.login_status == False:
      print(f"Please login to advise courses!")
    elif len(courses)==0:
      print(f"You haven't selected any courses.")
    elif len(courses)>3:
      print(f"You need special approval to take more than 3 courses.")
    else:
      print(f"Advising successful!")
      student.courses = courses
  def individualDetails(self,student):
    return f"Name: {student.name}\nID: {student.ID}\nDepartment: {student.department}\nAdvised courses: {(',').join(student.courses)}"

#Driver code
rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110", "CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))
```

```python
#Task 5
#Class Code
class Train:
  def __init__(self,name,*destinations):
    self.name = name
    self.start = destinations[0]
    self.end = destinations[-1]
    self.destinations = destinations
    self.passengers = []

    print(f"Welcome aboard on {self.name}")
    print(f"Start: {self.start}")
    print(f"Destination: {self.end}")

  def addPassenger(self,*passenger):
    for i in passenger:
      self.passengers.append(i)
      print(f"{i.name} welcome abroad")
      i.calculate_fare(self.destinations)

  def allPassengerDetails(self):
    for i in self.passengers:
      if i.start and i.end:
        print(f"Name: {i.name}, Start: {i.start}, Destination: {i.end}, Fair:{i.fare}")
      elif i.start:
        print(f"Name: {i.name}, Start: {i.start}, Destination: {self.end}, Fair:{i.fare}")
      elif i.end:
        print(f"Name: {i.name}, Start: {self.start}, Destination: {i.end}, Fair:{i.fare}")
      else:
        print(f"Name: {i.name}, Start: {self.start}, Destination: {self.end}, Fair:{i.fare}")

class Passenger:
  def __init__(self,name,start=None,end=None):
    self.name = name
    self.start = start
    self.end = end
    self.fare = 0
  def calculate_fare(self, stations):
    if self.start:
        start_index = stations.index(self.start)
    else:
        start_index = 0
    if self.end:
        end_index = stations.index(self.end)
    else:
      end_index = len(stations) - 1
    self.fare = 100 * abs(end_index - start_index)

#driver code
t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
#Task 6
#Class code
class BracuStudent:
  def __init__(self,name,address):
    self.name = name
    self.home = address
    self.bussPass = False

  def show_details(self):
    print(f"Student Name: {self.name}")
    print(f"Lives in {self.home}")
    print(f"Have buss pass? {self.bussPass}")

  def get_pass(self):
    self.bussPass = True

class BracuBus:
  def __init__(self,route,capacity=2):
    self.route =route
    self.capacity = capacity
    self.count = 0
    self.passengers = []

  def show_details(self):
    print(f"Bus Route: {self.route}")
    print(f"Passengers Count: {self.count} (Max:{self.capacity})")
    print(f"Passengers On Board: {self.passengers}")

  def board(self,*passengers):
    if passengers:
      for i in passengers:
        if i.bussPass and i.home == self.route:
          self.count+=1
          if self.count<=self.capacity:
            print(f"{i.name} boarded the bus.")
            self.passengers.append(i.name)
          else:
            self.count = self.capacity
            print(f"Bus is full!")
        elif i.home != self.route:
          print(f"You got on the wrong bus.")
        else:
          print(f"You don't have a buss pass.")
    else:
      print(f"No Passengers!")

#Driver code
st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()

```
