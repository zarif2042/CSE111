```python
#Task 1
def BMI(height,weight):
  height /= 100
  bmi = weight/(height**2)
  if bmi<18.5:
    print(f"Score is {round(bmi,1)}. You are underweight.")
  elif 18.5<=bmi<=24.9:
    print(f"Score is {round(bmi,1)}. You are normal.")
  elif 25<=bmi<=30:
    print(f"Score is {round(bmi,1)}. You are overweight.")
  elif bmi>30:
    print(f"Score is {round(bmi,1)}. You are obese.")
BMI(175,96)
BMI(152,48)
```

```python
#Task 2
def totalPrice(burger,place="Mohakhali"):
  meal_cost = 0
  delivery_charge = 40
  if burger == "BBQ Chicken Cheese Burger":
    meal_cost = 250
  elif burger == "Beef Burger":
    meal_cost = 170
  elif burger == "Naga Drums":
    meal_cost = 200
  if place != "Mohakhali":
    delivery_charge+=20
  total_cost = meal_cost+delivery_charge+meal_cost*8/100
  return total_cost
print(totalPrice("Beef Burger"))
```

```python
#Task 3
def replace_domain(email,new_domain,old_domain=None):
  new_email = ""
  if old_domain == None:
    return f"Unchanged: {email}"
  else:
    for value in email:
      new_email += value
      if value == "@":
        break
    new_email+=new_domain
    return f"Changed: {new_email}"
print(replace_domain("alice@kaaj.com","sheba.xyz","kaaj.com"))
print(replace_domain("bob@sheba.xyz","sheba.xyz"))
```

```python
#Task 4
def palindrome(string):
  is_palindrome = False
  new_string = ""
  for value in string:
    if value != " ":
      new_string += value
  for value1 in range(len(new_string)//2):
    if new_string[value1] != new_string[-(value1+1)]:
      print("Not a palindrome")
      break
  else:
    print("Palindrome")
palindrome("madam")
palindrome("hello")
palindrome("nurses run")
```

```python
#Task 5
def yearMonthDay(days):
  year = days//365
  days%=365
  month = days//30
  days%=30
  return f"{year} years, {month} months and {days} days"
print(yearMonthDay(4320))
print(yearMonthDay(4000))
```

```python
#Task 6
def title(string):
  new_string = ""
  for value3 in range(len(string)):
    if value3 == 0:
      new_string += chr(ord(string[value3])-32)
    else:
      new_string += string[value3]
  return new_string

def punctuation(string):
  special = [".",",","?","!"]
  flag = False
  lst = string.split(" ")
  new_string = ""
  for value in range(len(lst)):
    if value == 0 or flag:
      lst[value] = title(lst[value])
    if value == "i":
      lst[value] = "I"
    if value != len(lst)-1:
      x=lst[value]
      if x[-1] in special:
        flag = True
  for value1 in lst:
    new_string += value1+" "
  print(new_string)
punctuation("my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. do you know my pet dog's name? i love my pet very much.")
```
