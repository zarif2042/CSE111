```python
#Task 1
def hospital_fee(**kwargs):
    max_fee = max(kwargs.values())  # Find the maximum fee paid
    max_payers = [name for name, fee in kwargs.items() if fee == max_fee]  # List of names who paid the max fee

    return max_fee, max_payers

# Example usage:
max_amount, max_payer = hospital_fee(Mashrafe =
400, Bumrah = 900, Steyn = 1200, Cummins = 900,
Wood = 400, Marsh = 700)

(max_amount, ", ".join(max_payer))
```

```python
#Task 2
def is_james_bond(numbers):
    # Initialize an empty string to hold the pattern we are looking for
    pattern = ''

    # Loop through the numbers
    for num in numbers:
        # If the number is 0 and pattern is empty or has a single 0, add 0 to pattern
        if num == 0 and (pattern == '' or pattern == '0'):
            pattern += '0'
        # If the number is 7 and pattern has two 0s, add 7 to pattern and break the loop
        elif num == 7 and pattern == '00':
            pattern += '7'
            break

    # Check if the pattern is '007'
    return pattern == '007'

# Example usage:
example_1 = is_james_bond([1,2,4,0,0,7,5])  # Should return True
example_2 = is_james_bond([1,7,2,0,4,5,0])  # Should return False
example_3 = is_james_bond([1,0,2,0,4,7,5])  # Should return True

(example_1, example_2, example_3)
```

```python
#Task 3
def assign_students_to_sections(sections, *students):
    # Initialize a dictionary with empty lists for each section
    section_dict = {section: [] for section in sections}

    # Loop through the students
    for student in students:
        # Calculate the sum of ASCII values of the student's name
        ascii_sum = sum(ord(char) for char in student)

        # Determine the section using modulo operation
        section_index = ascii_sum % len(sections)
        assigned_section = sections[section_index]

        # Assign the student to the calculated section
        section_dict[assigned_section].append(student)

    return section_dict

# Example usage:
assign_students_to_sections('ABCDE', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace')
```

```python
#Task 4
def username_generator(first_name, last_name, student_id, middle_name=None):
    # First 3 letters of the first name upper-cased
    part_1 = first_name[:3].upper()

    # Entire middle name (if provided)
    part_2 = middle_name if middle_name is not None else ""

    # Last 3 letters of the last name lower-cased
    part_3 = last_name[-3:].lower()

    # Underscore
    part_4 = "_"

    # Last 4 digits of the student id
    part_5 = str(student_id)[-4:]

    # Concatenate all parts and return the username
    return part_1 + part_2 + part_3 + part_4 + part_5

# Example usage:
example_1 = username_generator('Jahanara', 'Islam', 18101954)  # Expected: JAHlam_1954
example_2 = username_generator('Shakib', 'Hasan', 17305165, 'Bin')  # Expected: SHABinsan_5165
example_3 = username_generator('MD', 'Ishmam', 19992564)  # Expected: MDmam_2564

(example_1, example_2, example_3)
```

```python
#Task 5
def key_generator(*names):
    encrypted_keys = []

    for name in names:
        part_1 = name[0].lower()

        part_3 = name[-1].upper()

        part_2 = ''.join(str(ord(char)) for char in reversed(name[1:-1]))

        encrypted_keys.append(part_1 + part_2 + part_3)

    return encrypted_keys

key_list = key_generator("Alex", "Bob", "Trudy")

print(key_list)
```

```python
#Task 6
import random

def playRockPaperScissor(rounds):
    # Possible moves in the game
    moves = ['rock', 'paper', 'scissor']

    # Initialize scores
    player_score = 0
    computer_score = 0

    # Game loop
    for _ in range(rounds):
        # Get player's move
        player_move = input("Enter your move (rock/paper/scissor): ").lower()

        # Get computer's move
        computer_move = random.choice(moves)
        print(f"Computer: {computer_move}")

        # Determine the winner of the round and update scores
        if player_move == computer_move:
            print("It's a tie!")
        elif (player_move == 'rock' and computer_move == 'scissor') or \
             (player_move == 'scissor' and computer_move == 'paper') or \
             (player_move == 'paper' and computer_move == 'rock'):
            print("You win this round!")
            player_score += 1
        else:
            print("Computer wins this round!")
            computer_score += 1

    # Display final scores
    print(f"\nYour Score: {player_score}")
    print(f"Computer's Score: {computer_score}")

    # Determine and display the overall winner
    if player_score == computer_score:
        print("The game is a tie!")
    elif player_score > computer_score:
        print("You have won the game!")
    else:
        print("Computer has won the game!")

# Example usage with 3 rounds (Uncomment to run interactively)
playRockPaperScissor(3)
```

```python
#Task 7
def capitalize_text(text):
    # Initialize variables
    new_text = ""
    capitalize_next = True
    prev_char = ""

    # Iterate through each character in the input string
    for char in text:
        # Handle multiple white spaces: If the current and previous characters are spaces, skip the loop iteration
        if char == " " and prev_char == " ":
            continue

        # Check if the next character should be capitalized
        if capitalize_next and char != " ":
            # Convert the character to uppercase if it's lowercase
            if 'a' <= char <= 'z':  # Check if char is a lowercase letter
                char = chr(ord(char) - ord('a') + ord('A'))  # Convert to uppercase
            capitalize_next = False  # Reset the flag

        # Check for conditions to capitalize the next character
        if char in [".", "!", "?"]:
            capitalize_next = True

        # Check for the lowercase "i" and capitalize it if it's isolated
        if char == "i" and (prev_char in [" ", ""] and new_text[-2:] in [" .", " !", " ?", ""]):
            char = "I"

        # Append the character to the new text
        new_text += char
        prev_char = char  # Update the previous character

    return new_text

# Test cases
sample_input_1 = ("my favourite animal is a dog.                a dog has sharp teeth so that it can eat flesh very easily. "
                  "do you know my pet dog's name? i love my pet very much.")
sample_input_2 = "i am a student.       i am stupid! am i?"

# Test the function
output_1 = capitalize_text(sample_input_1)
output_2 = capitalize_text(sample_input_2)

print(output_1)
```
