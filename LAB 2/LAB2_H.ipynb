```python
#Task 1
def hospital_fee(**kwargs):
    max_fee = float('-inf')
    max_p = []
    for value in kwargs.values():
       if value> max_fee:
            max_fee=value
    for key,value in kwargs.items():
        if kwargs[key] == max_fee:
            max_p.append(key)
    return (max_fee,max_p)
max_amount, max_payer = hospital_fee(Mashrafe =400, Bumrah = 900, Steyn = 1200, Cummins = 900, Wood = 400, Marsh = 700)
payers_string = ""
for value in range(len(max_payer)):
    if value != len(max_payer)-1:
        payers_string+=max_payer[value]+", "
    else:
        payers_string+=max_payer[value]+"."
print(f"Highest fee was {max_amount} tk which was paid by {payers_string}")
```

```python
#Task 3
def assign_students_to_sections(sections,*student_name):
    ascII_sum =0
    dict = {}
    for i in sections:
        dict[i]=[]
    section= 0
    for value in student_name:
        for value1 in value:
            ascII_sum += ord(value1)
        section = ascII_sum%len(sections)
        if sections[section] in dict:
            dict[sections[section]].append(value)
        ascII_sum = 0
        section = 0
    return dict
print(assign_students_to_sections('ABCDE', 'Alice','Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'))
```

```python
#Task 4
def username_generator(first_name,last_name,student_id,middle_name=""):
    student_id= str(student_id)
    username_string = ""
    if len(first_name)>3:
        username_string += first_name[:3].upper()
    else:
        username_string += first_name.upper()
    if middle_name:
        username_string += middle_name
    username_string+=last_name[-3:]+"_"+ student_id[-4:]
    return username_string
first_name, middle_name, last_name, student_id= input("First Name:"), input("Middle Name:"), input("Last Name:"), int(input("Student ID:"))
print(username_generator (first_name, last_name, student_id))
print(username_generator(first_name, last_name, student_id, middle_name))
```

```python
#Task 5
def key_generator(*args):
    new_l=[]
    for string in args:
        x = string[1:-1]
        key = ""
        key+= string[0].lower()
        for value in range(len(x)):
            key+=str(ord(x[-(value+1)]))
        key+= string[-1].upper()
        new_l.append(key)
        key=""
    return new_l
print(key_generator("Alex","Bob", "Trudy"))
```

```python
#Task 6
import random

def playRockPaperScissor(rounds):
    actions = ["rock", "paper", "scissors"]
    player_score = 0
    computer_score = 0

    for value in range(rounds):
        player_action = input("Enter your move (rock, paper, or scissors): ")
        computer_action = random.choice(actions)
        print(f"Computer: {computer_action}")

        if player_action == computer_action:
            print("It's a tie!")
        elif (player_action == "rock" and computer_action == "scissors") or (player_action == "paper" and computer_action == "rock") or (player_action == "scissors" and computer_action == "paper"):
            player_score += 1
            print("You win this round!")
        else:
            computer_score += 1
            print("Computer wins this round!")

    print(f"Your Score: {player_score}")
    print(f"Computer's Score: {computer_score}")

    if player_score > computer_score:
        print("You have won the game!")
    elif player_score < computer_score:
        print("Computer has won the game!")
    else:
        print("The game is a tie!")

playRockPaperScissor(3)
```
