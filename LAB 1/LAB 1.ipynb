```python
#Task 1
def upper(string):
    new_string = ""
    for value in string:
        if "A"<=value<="Z":
            new_string+=value
        elif "a"<=value<="z":
            new_string+=chr(ord(value)-32)
    return new_string

def lower(string):
    new_string = ""
    for value in string:
        if "a"<=value<="z":
            new_string+=value
        elif "A"<=value<="Z":
            new_string+=chr(ord(value)+32)
    return new_string

user_string = input()
up_count = 0
low_count = 0

for value in user_string:
    if "A"<= value <= "Z":
        up_count+=1
    elif "a"<= value <= "z":
        low_count+=1

if up_count>low_count:
    print(upper(user_string))
else:
    print(lower(user_string))
```

```python
#Task 4
def split(string,delimiter):
    lst = []
    n_str = ""
    index=0
    while index < len(string):
        if string[index:index + len(delimiter)] == delimiter:
            lst.append(n_str)
            n_str = ""
            index += len(delimiter)
        else:
            n_str += string[index]
            index += 1

    lst.append(n_str)
    return lst

is_common = False
user_string = input()
str1,str2 = split(user_string,", ")

for value in str1:
    if value in str2:
        print(value,end="")
        is_common = True

for value1 in str2:
    if value1 in str1:
        print(value1,end="")

if not is_common:
    print("Nothing in common")
```

```python
#Task 5
upper = False
lower = False
number = False
special = False
special_char = "_$#@"

user_string = input()

for value in user_string:
  if "A"<=value<="Z":
    upper = True
  if "a"<=value<="z":
    lower = True
  if "0"<= value <= "9":
    number = True
  if value in special_char:
    special = True

if not upper:
  print("Uppercase character missing")
if not lower:
  print("Lowercase character missing")
if not number:
  print("Digit missing")
if not special:
  print("Special character missing")
if upper and lower and number and special:
  print("OK")
```
