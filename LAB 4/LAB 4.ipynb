```python
#Task 1
#Class Code
class Student:
  def __init__(self,name,id,department="CSE"):
    self.name = name
    self.id = id
    self.department = department
    self.daily_effort = 0
  def dailyEffort(self,hour):
    self.daily_effort = hour
  def printDetails(self):
    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    print(f"Daily Effort: {self.daily_effort} hour(s)")

    if self.daily_effort <= 2:
      print(f"Suggestion: Should give more effort!")
    elif self.daily_effort<=4:
      print(f"Suggestion: Keep up the good work!")
    else:
      print(f"Suggestion: Excellent! Now motivate others.")

#Driver Code
harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
#Task 2
#Class Code
class Farmer:
  def __init__(self,name = None):
    if isinstance(name,int):
      print(f"Welcome to your farm. Your farm ID is {name}!")
    elif name:
      print(f"Welcome to your farm, {name}!")
    else:
      print(f"Welcome to your farm!")
    self.crops = []
    self.fishes = []

  def addCrops(self,*crops):
      for crop in crops:
        self.crops.append(crop)
      if self.crops:
        print(f"{len(crops)} crop(s) added.")
      else:
        print(f"No crop(s) added.")

  def addFishes(self,*fishes):
      for fish in fishes:
        self.fishes.append(fish)
      if self.fishes:
        print(f"{len(fishes)} fish(s) added.")
      else:
        print(f"No fish added.")

  def showGoods(self):
      if self.crops:
        print(f"You have {len(self.crops)} crop(s): ")
        print(f"{', '.join(self.crops)}")
      else:
        print(f"You don't have any crop(s)")
      if self.fishes:
        print(f"You have {len(self.fishes)} fish(s): ")
        print(f"{', '.join(self.fishes)}")
      else:
        print(f"You don't have any fish(s)")

#Driver Code
f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
#Task 3
#Class Code
class TaxiLagbe:
  def __init__(self,taxi_num,location):
    self.taxi_num = taxi_num
    self.location = location
    self.passangers = []
    self.fare = 0
  def addPassenger(self,*passangers):
    for passanger in passangers:
      if len(passangers)<=4:
        name,fare = passanger.split("_")
        self.passangers.append(name)
        self.fare += int(fare)
        print(f"Dear {name}! Welcome to TaxiLagbe.")
      else:
        print(f"Taxi Full! No more passengers can be added.")
        break

  def printDetails(self):
    print(f"Trip info for Taxi number: {self.taxi_num}")
    print(f"This taxi can only cover the {self.location} area.")
    print(f"Total passengers: {len(self.passangers)}")
    print(f"Passenger lists: ")
    print(f"{', '.join(self.passangers)}")
    print(f"Total collected fare: {self.fare} Taka")

#Driver Code
taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200','Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115', 'Parker_215')
print('-------------------------------')
taxi2.printDetails()

```

```python
#Task 4
#Class Code
class Sphere:
  def __init__(self,sphere_id,radius=1,color="White"):
    self.sphere_id = sphere_id
    self.radius = radius
    self.color = color
    self.volume = (4/3) * (3.1416) * (self.radius)**3

  def printDetails(self):
    print(f"Sphere ID: {self.sphere_id}")
    print(f"Color: {self.color}")
    print(f"Volume: {self.volume}")

  def merge_sphere(self,*sphere_ids):
    print("Spheres are being merged")
    for sphere_id in sphere_ids:
      self.volume += sphere_id.volume
      if self.color != sphere_id.color:
        self.color = "Mixed Color"


#Driver Code
sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()
print("2***************")
sphere2 = Sphere("Sphere 2", 3)
print("3***************")
sphere2.printDetails()
print("4***************")
sphere3 = Sphere("Sphere 3", 2)
print("5***************")
sphere3.printDetails()
print("6***************")
sphere3.merge_sphere(sphere1,sphere2)
print("7***************")
sphere3.printDetails()
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple")
print("9***************")
sphere4.merge_sphere(sphere3)
print("10***************")
sphere4.printDetails()

```
